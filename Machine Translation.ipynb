## PROBLEM 1 & 2 COMBINED

```python
## First get the unique terms
uni_de = []
uni_en = []
for i in range(len(sub_de)):
    g = sub_de[i]['de'].split(' ')
    e = sub_de[i]['en'].split(' ')
    for j in range(len(g)):
        if g[j] not in uni_de:
            uni_de.append(g[j])
    for j in range(len(e)):
        if e[j] not in uni_en:
            uni_en.append(e[j])
```

```python
## Creating dataframe for the matrix of unique terms
d = np.ones(shape=(len(uni_de),len(uni_en)))
t = pd.DataFrame(d, columns=uni_en, index=uni_de)
for i in range(len(uni_de)):
    for j in range(len(uni_en)):
        t.iloc[i][j] = t.iloc[i][j]/len(uni_en)
```

```python
## Calculating count and the Probabilities for each combination of unique words
for loop in range(50):
    d = np.zeros(shape=(len(uni_de),len(uni_en)))
    count = pd.DataFrame(d, columns=uni_en)
    total = np.zeros(len(uni_de))
    s_total = dict()
    for i in range(len(sub_de)):
        e = sub_de[i]['en'].split(' ')
        for j in e:
            for find in range(len(uni_en)):
                if j == uni_en[find]:
                    ei = find
            s_total[j] = 0
            g = sub_de[i]['de'].split(' ')
            for k in g:
                for find in range(len(uni_de)):
                    if k == uni_de[find]:
                        gi = find
                s_total[j] = s_total[j] + t.iloc[gi][ei]
        for j in e:
            for find in range(len(uni_en)):
                if j == uni_en[find]:
                    ei = find
            for k in g:
                for find in range(len(uni_de)):
                    if k == uni_de[find]:
                        gi = find
                count.iloc[gi,ei] = count.iloc[gi,ei] + t.iloc[gi,ei]/s_total[j]
                total[gi] = total[gi] + t.iloc[gi,ei]/s_total[j]
    for i in range(len(uni_de)):
        for j in range(len(uni_en)):
            t.iloc[i][j] = count.iloc[i][j]/total[i]
t = np.round(t, decimals=3)
```

```python
## Using EM Algorithm for word probabilities and IBM for allignment mapping

maximums = t.idxmax(axis=0)
mapping = dict()
for i in uni_en:
    mapping[i] = maximums[i]
for i in range(len(sub_de)):
    g = sub_de[i]['de'].split()
    e = sub_de[i]['en'].split()
    l1 = []
    l2 = []
    for j in e:
        l1.append(mapping[j])
    for j in l1:
        flag = 0
        for find in range(len(g)):
            if g[find] == j:
                flag = 1
                l2.append(find+1)
        if flag == 0:
            l2.append(0)
    for i in range(len(l2)):
        if l2[i] != 0:
            print('{}->{}'.format(i+1,l2[i]))
    print('\n')

print(mapping)
print(t)
```
